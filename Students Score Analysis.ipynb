```python
group_by = df.groupby("ParentEduc").agg({"MathScore": 'mean', "ReadingScore":'mean', "WritingScore": 'mean'})
print(group_by)
```

```python
sns.heatmap(group_by, annot = True)
plt.title("Relationship between Parent's Education Status and Student's Score")
plt.show()
```

### From the above chart we have got to no that, Education of Parents have a good impact on kids Score

## Distribution of Ethnic Groups

```python
groupA = df.loc[(df['EthnicGroup'] == "group A")].count()
groupB = df.loc[(df['EthnicGroup'] == "group B")].count()
groupC = df.loc[(df['EthnicGroup'] == "group C")].count()
groupD = df.loc[(df['EthnicGroup'] == "group D")].count()
groupE = df.loc[(df['EthnicGroup'] == "group E")].count()
lists = ["groupA", "groupB", "groupC", "groupD", "groupE"]
my_list = [groupA["EthnicGroup"], groupB["EthnicGroup"], groupC["EthnicGroup"], groupD["EthnicGroup"], groupE["EthnicGroup"]]
plt.pie(my_list, labels = lists, autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Group")
plt.show(lists)
```
